````python
# 1. Write a program that gets three numbers from the user and displays
# the biggest number among the three.    
def biggest_num(num_nums=3):
    nums = []
    while len(nums) < num_nums:
        new_num = input("Input a number...")
        nums.append(new_num)
    biggest = max(nums)
    print("Biggest number: ", biggest)
    return biggest

biggest_num()
````

````python
# 2. Write a program that gets a number from the user then says whether
# the number is an even number or odd number.

def even_or_odd(num=None):
    if num == None:
        num = input("Input a number...")
    oddity = int(num) // 2 != 0
    if oddity:
        print("Odd!")
        return "Odd"
    else:
        print("Even!")
        return "Even"
    
even_or_odd()
````

````python
# 3. Write a program that takes three numbers from the user as the lengths
# of three lines, then determines if the three lines can make a triangle.
# 4. add equilateral & isosceles checks
# 5. add right triangle check
# triangle rule: a + b > c, a + c > b, b + c + a
def can_it_triangle():
    A = float(input("Input line 1..."))
    B = float(input("Input line 2..."))
    C = float(input("Input line 3..."))
    check1 = A + B >= C
    check2 = A + C >= B
    check3 = B + C >= A
    if check1 and check2 and check3:
        if A == B and B == C:
            print("Equaliteral triangle")
        elif A**2 + B**2 == C**2:
            print("Right triangle")
        elif A == B or B == C:
            print("Isosceles triangle")
        else:
            print("some kind of triangle!")
    else:
        print("No triangle possible.")

can_it_triangle()
````

````python
# 6. Compound interest is a common practice in finance and banking,
# allowing you to earn interest on interest as well as on the principal.
# Assume that your bank offers a savings account with which you can
# earn compound interest. The amount you deposit into the account is
# p, and annual compound interest is r. By the end of n years after your
# initial deposit, the account balance will be a = p(1 + r)n. For this project,
# write a program that takes three numbers from the user as the initial
# deposit, the annual interest, and the number of years that the user
# wants the money to stay in the account. Calculate and display how
# much money the user will have by the end of the nth year.

def annual_compound_interest_value():
    p = float(input("Enter the principal investment..."))
    r = float(input("Enter the annual interest rate as a decimal, eg 5% = 0.05..."))
    n = float(input("Enter the number of years invested..."))

    final = p * (1 + r) ** n
    print("Final amount: ", final)

annual_compound_interest_value()
````

````python
# 7. In some countries like Canada, tax on taxable personal income for a
# year is calculated progressively according to a calculation table set for
# the year, such as the one shown below:
# 15% on $0–$48,534
# 20.5% on $48,536–$97,069
# 26% on $96,070–$150,473
# 29% on $150,474–$214,368
# 33% on $214,368 +
# Write a program that takes taxable income from a user and calculates
# the total income tax payable according to the table above.

def income_tax_calc(income):
    brackets = [214368, 150437, 97069, 48535, 0]
    rates = [.33, .29, .26, .205, .15]
    total_tax = 0
    for i in range(len(brackets)):
        bracket_top = income if i == 0 else brackets[i-1]
        applic_income = min(income, bracket_top) - brackets[i]
        print("applicable income: ", applic_income)
        if applic_income > 0:
            bracket_tax = applic_income * rates[i]
            total_tax += bracket_tax
            print(bracket_tax, " tax added")
    return total_tax


def calc_income_tax():
    income = float(input("Input total income..."))
    income_tax = income_tax_calc(income)
    print("income tax calculated: ", income_tax)
    
calc_income_tax()
````

````python
# 8. A mortgage is the money borrowed from a lender for the purchase
# of a property. Mortgages and mortgage payments are a big thing for
# almost everyone. For a mortgage with principal of P at a fixed monthly
# interest rate of r that needs to be paid off in Y years or 12 * Y = N
# months, the monthly payment would be:
# M = (rP(1+r)^n) / ((1+r)^n - 1)
# Write a program that takes the principal, fixed annual interest rate,
# and years of amortization then calculates and displays the monthly
# payment amount. Hint: You will need to work out the number of
# months from number of years, and the monthly interest rate from the
# annual interest rate.

def mortgage_payment_monthly(principal, annual_interest, years):
    r = annual_interest / 12 # fixed annual rate, not compounding monthly
    n = years * 12
    p = principal
    monthly_payment = (r * p * ((1 + r) ** n)) / (((1 + r) ** n) - 1 )
    return monthly_payment

print(mortgage_payment_monthly(400000, 0.028, 30))
````
