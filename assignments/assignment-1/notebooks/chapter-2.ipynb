````python
# 4 "generate *from* users?"
import random

def rand_user_mean(users = ["Neddie", "Bloodnok", "Eccles"]):
    random.seed(str(users))
    lower = ord(str(users[0][0])) * -1
    upper = ord(str(users[-1][-1]))
    nums = [random.randint(lower, upper) for u in range(len(users))]
    mean = sum(nums) / len(nums)
    print("="*79)
    print(f'Users: {", ".join(users)}.')
    print(f'\nGenerated bounds: (-){abs(lower)} ("[{users[0][0]}]{users[0][1:]}" - "{users[0][0]}") to {upper} ({users[-1][:-1]}[{users[-1][-1]}] - "{users[-1][-1]}")')
    print(f'\nGenerated random integers: {nums}')
    print(f'\nMean: {mean}.  \n\nNeat!')
    
rand_user_mean()
````

````python
# 5
def cube_SA(edge = input("Input edge length...")):
    sa = 6 * float(edge) ** 2
    print("Cube surface area: ", sa)
    return sa

cube_SA()
````
