### Project 1
a. significant computing machines
- abacus - manual, slow (but faster than other options), still huge upgrade from mental math or counting
- difference engine and analytical engine (charles babbage) - mechanical computation, stored programs
- collossus mk 1 and 2 - enigma code cracking, more powerful, specialized
- ENIAC - U of Penn, general purpose electronic computer (18000+ tubes!)

b. key contributors
- Charles Babbage - designed the first mechanical computers
- George Boole - developed boolean algebra
- Alan Turing - theoretical basis of computation (limits - what is or isn't computable)
- John van Neumann described arch for stored programs that is still used
- Ada Lovelace - wrote first program for Babbage analytical engine

c. Key tech/concepts
- boolean algebra - logical foundation of binary computing
- stored-program concept - allows computers to to both store instructions in memory and execute stored instructions
- vacuum tubes/transistors - allow electronic representation of binary computing
- ICs - miniturization of transistors


### Project 2
a. notable machine languages
- intel x86 - basis of PCs and widely used in desktops and servers
- ARM - dominates mobile/embedded systems
- RISC-V - of interest to me due to improvements in ESP32 devices, specifically ESP32-C3, which is fully RISC-V instead of including an Xtensa LX6/LX7 main/coprocessor

b. machine vs assembly vs high-level languages
- machine language - binary (0/1) directly executable by processor. Each processor archtecture has its own machine language.
- assembly language - human readable mnemonics (eg ADD, MOV) mapped to machine language via an "assembler" translator program.
- high-level language - hardware details abstracted away for readability and portability, compiled or interpreted to machine code.

c. structured programming 
- clearly defined control structures, sequence/conditional/loop. 
- avoids unstructured jumps.
- readability, debugging, maintenance good; can be very verbose depending on problem type
- C, Pascal

d. imperative programming
- sequence of statements that change state - telling *how* to do something step by step.
- similar to how hardware works, can be intuitive, performant; can be difficult to manage at scale
- C, Pascal, Fortran, Python

e. declarative programming
- describes desired outcome, not steps
- concise, easy to reason about
- limited control over performance
- SQL, HTML, Prolog

f. functional programming
- accomplishes computation as evaluation of mathematical functions, avoiding mutale state
- good for parallelism, few side-effect bugs, makes emacs exist (which is good)
- some performance overhead (sometimes), can be very unfamiliar for imperative programmers
- lisp, haskell, erlang, some parts of python and JS

g. logic programming
- expresses process as logical facts ad rules, "inference engine" derives conclusions
- powerful for constraint solving and GOFAI
- limited general use, can be slow
- prolog, datalog

h. object-oriented programming
- organizes code as objects with attributes and behavior, using inheritance, polymorphism, encapsulation
- models real-world entities naturally, modularity and re-use are easy
- inheritance heirachies can be brittle, can lead to over-engineering
- Java, C++, Python, C#
